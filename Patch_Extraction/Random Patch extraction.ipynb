<h3>Random Patch Extraction</h3>

```python
import os
import random
import shutil


def is_destination_directory_exists(destination):
    if not os.path.exists(destination):
        os.mkdir(destination)
    elif os.path.exists(destination):
        pass
    
    return


def random_sample_patches(files,size):
    
    if len(files) < size:
        samples = random.sample(files,len(files))
    else:
        samples = random.sample(files,size)
        
    return samples

def make_directory(destination,name):
    
    if not os.path.exists(destination+name):
        os.mkdir(destination+name)
    else:
        pass
                      
```

```python
"""Please specify set_segment - Train/Test/Validation

Classification - creates a folder and extract the patches from all the whole slide images based on the classification

destnation - it is destination path to save the patches

source_path -  self explanatory( a source path)"""
```

```python
"""
folders structure 
Dataset/ Train & Test & Validation folders
Each folders has sub folders of Whole slide images and it randomly sampled patches 
"""
```

```python
set_segment = 'Validation'
classification = 'Nonepithelioid'
size = 100

destination = 'Desktop/new_dataset/Dataset_100/'+set_segment+'/'+classification+'/'
source_path ='/Volumes/USB/All_dense_patches/'+set_segment+'/'+classification+'/'

is_destination_directory_exists(destination)

for dirs,sub,files in os.walk(source_path):
    for name in sub:
        for dirs,sub,files in os.walk(source_path+name):
            
            file_samples_list = random_sample_patches(files,size)
                
            make_directory(destination,name)
            
            for i in file_samples_list:
                shutil.copy(source_path+name+'/'+i, destination+name)
```
